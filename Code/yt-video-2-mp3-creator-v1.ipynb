```python
def get_youtube_title(url):
    """
    Retrieves the title of a YouTube video using yt-dlp.

    Args:
        url: The URL of the YouTube video.

    Returns:
        The title of the video as a string, or None if an error occurs.
    """
    try:
        # Use yt-dlp to get video information
        command = f"yt-dlp -j '{url}'"  # Use -j for JSON output
        result = os.popen(command).read()
        import json
        video_info = json.loads(result)
        title = video_info.get('title')
        return title
    except Exception as e:
        print(f"Error getting video title: {e}")
        return None
```

```python
import os

def download_youtube_audio(url, output_path=".", output_file="output.mp3"):
    """Downloads the audio of a YouTube video as an MP3 file.

    Args:
        url: The URL of the YouTube video.
        output_path: The directory to save the downloaded MP3 file.
    """

    try:
        # Use yt-dlp to download the audio
        download_command = f"yt-dlp -x --audio-format mp3 -o '{output_path}/%(title)s.%(ext)s' {url}"

        os.system(download_command)

        print(f"Download complete. Audio saved to {output_path}")

        title = get_youtube_title(url)
        output_file_old_name = os.path.join(output_path, title + ".mp3")
        output_file_new_name = os.path.join(output_path, output_file)

        print(f"Old file name: {output_file_old_name}")
        print(f"New file name: {output_file_new_name}")

        os.rename(output_file_old_name, output_file_new_name)

        print(f"Created audio file {output_file_new_name}")

    except Exception as e:
        print(f"An error occurred during download: {e}")
```
